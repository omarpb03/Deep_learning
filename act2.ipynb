```python
# Cálculo de riesgo empírico que corresponde a un función de pérdida
import numpy as np
import pandas as pd
```

```python
# Datos tabla 1
x = np.array([-2, -1, 0, 1, 2])
y = np.array([3, 2, 1, 1, 0.5])
```

```python
# Funciones de pérdida
f1 = 1.5 - 0.6 * x
f2 = 1.5 - 0.5 * x

# Pérdida cuadrática
loss_quad_f1 = (y - f1) ** 2
loss_quad_f2 = (y - f2) ** 2

# Pérdida absoluta
loss_abs_f1 = np.abs(y - f1)
loss_abs_f2 = np.abs(y - f2)

# Funciones
f1 = 1.5 - 0.6 * x
f2 = 1.5 - 0.5 * x
```

```python
# Riesgos empíricos
print("Riesgo empírico")
print("Pérdida cuadrática:")
print(f"Función 1: {loss_quad_f1.mean():.4f}")
print(f"Función 2: {loss_quad_f2.mean():.4f}")

print("\nPérdida absoluta:")
print(f"Función 1: {loss_abs_f1.mean():.4f}")
print(f"Función 2: {loss_abs_f2.mean():.4f}")
```
